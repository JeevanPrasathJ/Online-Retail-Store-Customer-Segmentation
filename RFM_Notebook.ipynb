# Customer Segmentation with RFM Scores
RFM is abbreviation for Recency, Frequency and Monetary. It is a technique that helps determine marketing and sales strategies based on customers' buying habits. Recency: Time passed since the customer's last purchase. In other words, it is the "time since the last contact of the customer"

```python
retaildata_2.head()
```

```python
retaildata_2.info()
```

```python
retaildata_2["InvoiceDate"].min()
```

```python
retaildata_2["InvoiceDate"].max()
```

Now if we take date as the analyse date, then there will be a very serious difference.


For this reason, we have set the next day of the maximum day of the data set as the analyse date

```python
import datetime as dt
Last_date = dt.datetime(2010,12,10)
Last_date
```

```python
retaildata_2.groupby("Customer ID").agg({"InvoiceDate":"max"}).head(10)
```

what should we do now?

For each customer, we need to subtract the customers' last purchase date from this date.
And we will keep this as a temp data frame.

```python
(Last_date - retaildata_2.groupby("Customer ID").agg({"InvoiceDate":"max"})).head(10)
```

```python
temp_retaildata = (Last_date - retaildata_2.groupby("Customer ID").agg({"InvoiceDate":"max"}))
```

```python
(Last_date - retaildata_2.groupby(["Customer ID","Description"]).agg({"InvoiceDate":"max"})).head(60)
#I did this to check whether the Customer IDs are valid IDs with valid transactions
```

```python
#changing the name of InvoiceDate to Recency
temp_retaildata.rename(columns={"InvoiceDate": "Recency"}, inplace = True)
```

# Frequency

First, by grouping the data based on Customer ID and Invoice Number, we observe how many times each invoice is duplicated for each customer. Let's take a look at the first 20 observations.

```python
temp_retaildata = retaildata_2.groupby(["Customer ID","Invoice"]).agg({"Invoice":"count"})
temp_retaildata.head(20)
```

# Monetary



Monetary is the total spend of the customer.

The sum of the TotalPrice values that we have calculated for each customer before is that customer's monetary value:

```python
monetary_retaildata = retaildata_2.groupby("Customer ID").agg({"TotalPrice":"sum"})
monetary_retaildata.head()
```

```python
# Finding the customer which we gained the most money from: 18102 (the customer with most orders is 14911 so they are not the same)
monetary_retaildata.sort_values(by="TotalPrice", ascending=False).head(10)
```

```python
monetary_retaildata.rename(columns={"TotalPrice": "Monetary"}, inplace = True)
monetary_retaildata.head(10)
```

```python
print(recency_retaildata.shape, freq_retaildata.shape, monetary_retaildata.shape)
```

```python
#Lets bring all the values together under a new data frame called 'rfm':
rfm = pd.concat([recency_retaildata, freq_retaildata, monetary_retaildata],  axis=1)
rfm.head(10)
```

Normally the smallest of the recency scoring, which is 1, is the best recency score.

However, we will define this in reverse and put the value 5 as the best recency value so that it will be the same as the others,
so score 5 will be the most recent and the best recency score:

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 5, labels = [5, 4, 3, 2, 1])
```

```python
rfm["Frequency"].describe([0.01,0.05,0.10,0.25,0.50,0.60,0.75,0.90,0.95, 0.99]).T
```

```python
cut_bins = [0,1,2,3,9,205]

rfm["FrequencyScore"] = pd.cut(rfm["Frequency"], bins = cut_bins, labels = [1, 2, 3, 4, 5])
rfm["FrequencyScore"].value_counts()
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
rfm.head(10)
```

```python
(rfm['RecencyScore'].astype(str) +
 rfm['FrequencyScore'].astype(str) +
 rfm['MonetaryScore'].astype(str)).head(10)
```

```python
rfm["RFM_SCORE"] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str) + rfm['MonetaryScore'].astype(str)
rfm.head()
```

```python
#To segment the customers using Recency and Frequency values:
#set up regular expression (regex) structure by using dictionaries to name Customer Segments according to
#Recency and Frequency Scores:

seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Lose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
#recency ve frequency points are obtained with the following code
#(as the monetary is a similar value to frequency, it is not considered here):
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'].head()
```

```python
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
#Retrieve statistical values of recency, frequency and monetary values by segment classes:
rfm[["Segment", "Recency","Frequency","Monetary"]].groupby("Segment").agg(["mean","count"])
```
